```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from google.colab import drive
import pandas as pd

# Google Drive'ı bağla
drive.mount('/content/drive')
```

```python
# Dosyanın yolunu belirle
file_path = '/content/drive/My Drive/Hitters.csv'  # Dosyanın tam yolu

# Veriyi oku
df = pd.read_csv(file_path)
df.head()
```

```python
df = df.dropna()
dms = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
y = df["Salary"]
X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
svr_model = SVR(kernel="linear")
svr_model.fit(X_train, y_train)
```

```python
print(svr_model.coef_)
print(svr_model.intercept_)
```

```python
svr_model.predict(X_test[0:5])
```

```python
y_pred = svr_model.predict(X_test)
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
svr_model2 = SVR(kernel="rbf")
svr_model2.fit(X_train, y_train)
```

```python
y_pred2 = svr_model2.predict(X_test)
np.sqrt(mean_squared_error(y_test, y_pred2))
```

```python
SVR_params = {"C": [0.1, 0.5, 0.7, 1.4, 1.7]}
```

```python
SVR_CV_Model = GridSearchCV(svr_model2, SVR_params, cv=5).fit(X_train, y_train)
```

```python
SVR_CV_Model.best_params_
```

```python
SVR_CV_Model = GridSearchCV(svr_model2, SVR_params, verbose=2, n_jobs = -1).fit(X_train, y_train)
```

```python
SVR_CV_Model.best_params_
```
