```python
# Dosyanın yolunu belirle
file_path = '/content/drive/My Drive/Hitters.csv'  # Dosyanın tam yolu

# Veriyi oku
df = pd.read_csv(file_path)
df.head()
```

K Nearest Neighbors

```python
df = df.dropna()
dms = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
y = df["Salary"]

X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2, random_state=45)
```

```python
KnnModel = KNeighborsRegressor().fit(X_train, Y_train)
```

```python
KnnModel
```

```python
KnnModel.n_neighbors
```

```python
KnnModel.predict(X_test[0:10])
```

```python
y_pred = KnnModel.predict(X_test)
```

```python
np.sqrt(mean_squared_error(Y_test, y_pred))
```

Hand

```python
RMSE = []
k = 1
for i in range(10):
  k = k + 2
  KnnModel = KNeighborsRegressor(n_neighbors=k).fit(X_train, Y_train)
  y_pred = KnnModel.predict(X_train)
  rmse = np.sqrt(mean_squared_error(Y_train, y_pred))
  RMSE.append(rmse)
  print("k=", k, "için RMSE değeri:", rmse)
```

Auto // GridSearchCV

```python
knnParams = {"n_neighbors": np.arange(1, 20, 2)}
KnnModel2 = KNeighborsRegressor()
knnCvModel = GridSearchCV(KnnModel2, knnParams, cv=10).fit(X_train, Y_train)
```

```python
knnCvModel.best_params_
```

```python
KnnModel = KNeighborsRegressor(n_neighbors= 7).fit(X_train, Y_train)
y_pred = KnnModel.predict(X_test)
np.sqrt(mean_squared_error(Y_test, y_pred))
```
